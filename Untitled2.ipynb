## Request places

```python
places_request = {}
for i in node_1_data["places"]:
    for j in i:
        if not (j in list(places_request.keys())):
            print(f"Request: {j}")
            url_request = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={j}&inputtype=textquery&fields=photos,formatted_address,name,geometry&key={api_key}"
            places_request[j] = requests.get(url_request)
```

```python
places_data = {}
for i in places_request:
    place_info = places_request[i].json()
    #print(place_info)
    print(i)
    print(len(place_info['candidates']))
    if place_info['candidates']:
        #print(place_info['candidates'][0])
        places_data[i] = {
            "lat": place_info['candidates'][0]["geometry"]["location"]["lat"],
            "lng": place_info['candidates'][0]["geometry"]["location"]["lng"],
            "photo": place_info['candidates'][0]["photos"][0]["photo_reference"] if "photos" in place_info['candidates'][0] else "",
        }
```

```python
places_df = pd.DataFrame(places_data).T
places_df["Name"] = places_df.index
```

```python
fig = gmaps.figure()
```

```python

initial_node_point = (19.3328625,-99.2440273)

points_layer = gmaps.symbol_layer(
    places_df[["lat","lng"]], fill_color=(253,163,15), stroke_color=(253,163,15), scale=2
)
initial_node_layer = gmaps.symbol_layer(
    [initial_node_point], fill_color=(2,20,40), stroke_color=(2,20,40), scale=3
)


drawing = gmaps.drawing_layer(features=[gmaps.Line((places_df.iloc[i]["lat"], places_df.iloc[i]["lng"]), initial_node_point, stroke_weight=2.0,stroke_color=(27,77,100)) for i in range(len(places_df.index))])
fig.add_layer(drawing)
fig.add_layer(points_layer)
fig.add_layer(initial_node_layer)
fig
```
